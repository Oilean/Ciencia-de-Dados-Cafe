# Importando os dados e a tabela

```python
import pandas as pd

url = "https://gist.githubusercontent.com/Oilean/ff04ca0010125904266d162a28dc23b2/raw/363db91a67442a831946507cc6e6e76eda68b575/gistfile1.txt"
tabela = pd.read_csv(url)

tabela.head()
```

# Configurando a tabela

```python
# Verificando se há valores nulos na tabela
tabela.info()
```

```python
data = tabela.reset_index(drop=True)
n=len(data)
print(n)
display(data)
```

```python
data['Year']=data['Date'].apply(lambda x:int(x[0:4]))
data['Month']=data['Date'].apply(lambda x:int(x[5:7]))
data['Year-Month']=data['Date'].apply(lambda x:x[0:7])
data['Date']=pd.to_datetime(data['Date'])
display(data)
```

```python
data.describe()
```

# Importando as bibliotecas gráficas

```python
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import seaborn as sns
```

<h1> Gráficos Reproduzidos

<h3> Preço Diário do Café

Abaixo mostra-se a variação do preço do café, de acordo com o volume:

```python
tabela["Vol"] = tabela["Volume"]/100

tabela["Open"].plot(label = "Abertura", color = "#06BE2D", linewidth = 5, figsize = (13,5))
tabela["Close"].plot(label = "Fechamento", color = "#F91D1D", linewidth = 0.75, figsize = (13,5))
tabela["Vol"].plot(label = "Volume", color = "#101CCB", alpha = 0.25, linewidth = 1, figsize = (13,5))

plt.grid(axis = "y", color = "#000000")
plt.title("Preço Diário do Café")
plt.xlabel("Abertura Diária do Mercado Desde 2000")
plt.ylabel("USD")
plt.legend(loc = 2)
plt.show()
```

<h3>Média de Volume de Café Vendido por Ano

No gráfico abaixo, pode-se visualizar, o aumento da venda de cafés ao decorrer dos anos:



```python
dataPorAno = data.groupby(["Year"]).mean('Volume')

data1 = dataPorAno.index
data2 = dataPorAno['Volume']

plt.style.use('ggplot')

plt.figure(figsize=(10,8))
plt.barh(data1,data2)
plt.title('Média de Volume de café vendido por ano')
plt.ylabel('Ano')
plt.xlabel('Volume médio')
plt.show()
```

#### Com isso, podemos partir para a preparação dos dados para treinarmos o Modelo de Machine Learning

- Separando em dados de treino e dados de teste

```python
y = tabela[['Close']]
x = tabela[['Open', 'Low', 'High', 'Volume']]

from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.2)
```

```python
print(f"Tamanho de x_treino: {x_treino.shape}")
print(f"Tamanho de x_teste: {x_teste.shape}")
print(f"Tamanho de y_treino: {y_treino.shape}")
print(f"Tamanho de y_teste: {y_teste.shape}")
```

- Criando a interligencia artificial e fazendo as previsões 

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

# cria as inteligencias aritificiais
modelo_regressaolinear = LinearRegression()
modelo_arvoredecisao = RandomForestRegressor()

# treina as inteligencias artificias
modelo_regressaolinear.fit(x_treino, y_treino)
modelo_arvoredecisao.fit(x_treino, y_treino)
```

```python
from sklearn import metrics

# criar as previsoes
previsao_regressaolinear = modelo_regressaolinear.predict(x_teste)
previsao_arvoredecisao = modelo_arvoredecisao.predict(x_teste)

# comparar os modelos
print(metrics.r2_score(y_teste, previsao_regressaolinear))
print(metrics.r2_score(y_teste, previsao_arvoredecisao))  
```

- Com isso vemos que nesse caso o Modelo de Regrassão Linear obteve melhor resultado

```python
# Criar o gráfico da previsão
NovaTabela = pd.DataFrame()

NovaTabela['y_teste'] = y_teste
NovaTabela['Previsões Regressão Linear'] = previsao_regressaolinear

plt.figure(figsize = (10, 8))
sns.lineplot(data = NovaTabela)
plt.show()
```
